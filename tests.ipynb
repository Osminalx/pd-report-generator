```python
files = [
    (
        "./Archivos/LAMINAS GALVANIZADOS RELACION DE SCRAP EN CONTENEDORES METAL-HULE.xlsm",
        "RELACION GALVANIZADO",
    ),
    (
        "./Archivos/LAMINAS ALUMINIOS RELACION DE SCRAP EN CONTENEDORES METAL-HULE.xlsm",
        "RELACION ALUMINIO",
    ),
    (
        "./Archivos/LAMINAS STAINLESS RELACION DE SCRAP EN CONTENEDORES METAL-HULE.xlsm",
        "RELACION STAINLESS",
    ),
]

dataframes = []

# Columnas necesarias
columnas_necesarias = ["Fecha", "# de Orden", "SCRAP/SHT"]

# Leer cada archivo y su hoja específica
for archivo, hoja in files:
    df = pd.read_excel(archivo, sheet_name=hoja, header=9)

    # Filtrar solo las columnas necesarias (verifica los nombres exactos)
    df = df[columnas_necesarias]

#     # Agregar el DataFrame a la lista
    dataframes.append(df)

# # Concatenar todos los DataFrames en uno solo
df_final = pd.concat(dataframes, ignore_index=True)
df_final
```

```python
files = [
    ("./Archivos/OMBLIGOS ALUMINIOS RELACION DE SCRAP EN CONTENEDORES METAL-HULE.xlsm","RELACION ALUMINIO"),
    ("./Archivos/OMBLIGOS GALVANIZADOS RELACION DE SCRAP EN CONTENEDORES METAL-HULE.xlsm","RELACION GALVANIZADO")
]

dataframes = []

needed_columns = [1,2,16]

for file,sheet in files:
    df = pd.read_excel(file,sheet_name=sheet,header=9)

    df = df.iloc[:,needed_columns]
    
    df.columns = ["Fecha","# de Orden","SCRAP/SHT"]

    dataframes.append(df)

final_df = pd.concat(dataframes,ignore_index=True)
final_df
```
